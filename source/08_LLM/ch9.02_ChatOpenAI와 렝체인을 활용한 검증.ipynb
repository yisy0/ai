# 2. LLM 답변 생성
## 2.1 OpenAI SDK를 선택
- pip install openai / pip install langchain-openai

```python
from openai import OpenAI
client = OpenAI()
response = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages=[
        {"role":"system", "content":"너는 친절하게 대답해 주는 비서야"},
        {"role":"user", "content":"2020년 월드 시리즈 누가 우승했어?"}
    ],
    temperature=0.5, # 0~0.3:일관되고 예측가능한답 / 0.8~1:창의적다양한단어답변
)
```
